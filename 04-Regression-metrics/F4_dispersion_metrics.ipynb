
El efecto de la dispersin en las mtricas

Veamos cmo los valores EAM y RECM dependen de la dispersin del objetivo.

A diferencia de EAM, la mtrica RECM es bastante sensible a valores grandes: los errores significativos afectan fuertemente el valor final de RECM.

Echa un vistazo a los siguientes tres grficos de distribucin de errores (los valores de error se trazan a lo largo del eje horizontal y el nmero de errores est a lo largo del vertical).

En el primer grfico, el nmero de errores pequeos (con valores en el rango de 0 a 10) y el nmero de errores grandes (con valores de 40 y superiores) son iguales.
En el segundo y tercer grfico, la brecha aumenta gradualmente. Comienzan a aparecer errores muy grandes, aumenta el nmero de errores pequeos. Los valores de EAM no cambian: los errores grandes se equilibran con los errores pequeos. El RECM, por otro lado, aumenta, porque para RECM tomamos los errores al cuadrado y esto hace que la mtrica sea ms sensible a errores de gran magnitud, aunque dichos errores son pocos.



Ahora sabes que es posible cambiar el valor de una mtrica sin cambiar otra.
